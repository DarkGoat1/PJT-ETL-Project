# Windows Users

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
nuurl = 'https://www.smogon.com/dex/gs/formats/nu/'
browser.visit(nuurl)
```

```python
import re

html = browser.html
soup = BeautifulSoup(html, 'html.parser')

s = soup.find_all("div", href=re.compile("/dex/gs/formats/"))
nutiers = soup.find_all("a", href=lambda href: href and "/dex/gs/formats/" in href)

nutiers
```

```python
uuurl = 'https://www.smogon.com/dex/gs/formats/uu/'
browser.visit(uuurl)

html = browser.html
soup = BeautifulSoup(html, 'html.parser')

s = soup.find_all("div", href=re.compile("/dex/gs/formats/"))
uutiers = soup.find_all("a", href=lambda href: href and "/dex/gs/formats/" in href)

uutiers
```

```python
uublurl = 'https://www.smogon.com/dex/gs/formats/uubl/'
browser.visit(uublurl)

html = browser.html
soup = BeautifulSoup(html, 'html.parser')

s = soup.find_all("div", href=re.compile("/dex/gs/formats/"))
uubltiers = soup.find_all("a", href=lambda href: href and "/dex/gs/formats/" in href)

uubltiers
```

```python
ouurl = 'https://www.smogon.com/dex/gs/formats/ou/'
browser.visit(ouurl)

html = browser.html
soup = BeautifulSoup(html, 'html.parser')

s = soup.find_all("div", href=re.compile("/dex/gs/formats/"))
outiers = soup.find_all("a", href=lambda href: href and "/dex/gs/formats/" in href)

outiers
```

```python
alltiers = nutiers + uutiers + uubltiers + outiers
```

```python
import pandas as pd


tierframe = pd.DataFrame(data=alltiers)   

tierframe.to_csv('tierinfo.csv')
```
